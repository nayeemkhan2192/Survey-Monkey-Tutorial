```python
import pandas as pd
import numpy as np
import io

```

```python
from google.colab import files
uploaded =files.upload()
```

```python
dataset = pd.read_excel(io.BytesIO(uploaded['data_actual.xlsx']))
dataset
```

```python
dataset_modified = dataset
dataset_modified
```

```python
dataset_modified.columns
```

```python
columns_to_drop= [ 'Start Date-', 'End Date-', 'Email Address-','First Name-', 'Last Name-', 'Custom Data 1-']
columns_to_drop
```

```python
dataset_modified= dataset
dataset_modified = dataset_modified.drop(columns=columns_to_drop)
dataset_modified
```

```python
id_vars=list(dataset_modified.columns)[:8] 
value_vars = list(dataset_modified.columns)[8 :] 
#value_vars
```

```python
dataset_melted = dataset_modified.melt(id_vars=id_vars,value_vars=value_vars,var_name='Question_SubQuestion',value_name='Answers')
dataset_melted
```

```python
from google.colab import files
uploaded2 =files.upload()
```

```python
questions_import = pd.read_excel(io.BytesIO(uploaded2['question.xlsx']))
questions_import
```

```python
questions = questions_import.copy()
questions
```

```python
questions.columns
questions=questions.drop(columns= ['RawResponds', 'Questions', 'Subquestion'])
```

```python
questions.dropna(inplace=True)
questions
```

```python
questions.columns
```

```python
question2 = questions.rename(columns={'Question+Subquestion':'Question_SubQuestion'},inplace=False)
question2
```

```python
dataset_merged = pd.merge(left= dataset_melted,right= question2 ,how="left",left_on="Question_SubQuestion",right_on="Question_SubQuestion")
print("Original Data",len(dataset_melted))
print("Original Data",len(dataset_merged))
```

```python
dataset_merged
```

```python
dataset_merged.groupby("RawQuestion")["Respondent ID-"].count().reset_index()
```

```python
respondents = dataset_merged[dataset_merged["Answers"].notna()]
respondents = respondents.groupby("RawQuestion")["Respondent ID-"].nunique().reset_index()
respondents.rename(columns={"Respondent ID-":"Respondents"}, inplace=True)
respondents
```

```python

dataset_merged_two = pd.merge(left=dataset_merged, right=respondents, how="left", left_on="RawQuestion", right_on="RawQuestion")
print("Original Data", len(dataset_merged))
print("Merged Data", len(dataset_merged_two))
dataset_merged_two
```

```python
same_answer = dataset_merged
same_answer = same_answer.groupby(["Question_SubQuestion","Answers"])["Respondent ID-"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID-":"Same Answer"}, inplace=True)
same_answer
```
